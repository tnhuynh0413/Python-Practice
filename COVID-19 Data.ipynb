```python
# import libraries

import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
corona_dataset_csv = pd.read_csv("/Users/tienhuynh/Downloads/covid19_Confirmed_dataset.csv") # import dataset
corona_dataset_csv.head(10)
```

```python
corona_dataset_csv.shape
```

```python
corona_dataset_csv.drop(["Lat","Long"], axis = 1, inplace = True) # delete columns that are not needed
```

```python
corona_dataset_csv.head(10)
```

```python
corona_dataset_aggregated = corona_dataset_csv.groupby("Country/Region").sum() # aggregate rows by country
```

```python
corona_dataset_aggregated.drop(["Province/State"], axis = 1, inplace = True)
```

```python
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.shape
```

```python
# visualize data related to several selected countries

corona_dataset_aggregated.loc["China"].plot()
corona_dataset_aggregated.loc['Italy'].plot()
corona_dataset_aggregated.loc['Spain'].plot()
plt.legend()
```

```python
corona_dataset_aggregated.loc["China"].plot()
```

```python
# finding the first derivative for the curve (China)

corona_dataset_aggregated.loc["China"].diff().plot()
```

```python
# finding the maximum infection rate for China

corona_dataset_aggregated.loc["China"].diff().max()
```

```python
# finding the maximum infection rate for Italy

corona_dataset_aggregated.loc["Italy"].diff().max()
```

```python
# finding the maximum infection rate for Spain

corona_dataset_aggregated.loc["Spain"].diff().max()
```

```python
# finding the maximum infection rate for all the countries

countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
max_infection_rates
corona_dataset_aggregated["max_infection_rates"] = max_infection_rates
```

```python
corona_dataset_aggregated.head()
```

```python
corona_data = pd.DataFrame(corona_dataset_aggregated["max_infection_rates"]) # new df with max_infection_rates
```
